```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import seaborn as sns
import matplotlib.pyplot as plt
print("Import Successful")
```

```python
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Impute missing values
age_imputer = SimpleImputer(strategy='median')
embarked_imputer = SimpleImputer(strategy='most_frequent')

df['Age'] = age_imputer.fit_transform(df[['Age']])
df['Embarked'] = embarked_imputer.fit_transform(df[['Embarked']]).ravel()

# Drop 'Cabin' col
df.drop(columns=['Cabin'], inplace=True)

# Verify the changes
df.isnull().sum()
```

```python
df.head()
```

```python
df.describe()
```

```python
# Encode categorical variables
# One-Hot encoding for Sex and Embarked
encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_sex_embarked = encoder.fit_transform(df[['Sex', 'Embarked']])
encoded_df = pd.DataFrame(encoded_sex_embarked, columns=encoder.get_feature_names_out(['Sex', 'Embarked']))

# Drop original Sex and Embarked columns and concatenate the encoded cols
df.drop(columns=['Sex', 'Embarked'], inplace=True)
df = pd.concat([df, encoded_df], axis=1)

df.head()
```
